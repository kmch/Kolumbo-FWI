```python
import numpy as np
import matplotlib.pyplot as plt
from arrau.a2d import Arr2d
from fullwavepy.plot.misc import plot_square
from fullwavepy.seismic.misc import Box3d
from fullwavepy.seismic.proteus import PROTEUS
from plotea.mpl2d import figure
```

```python
xp = PROTEUS()
xp.read_all()
b01 = Box3d(+8.0e3, +2.5e4 , -3.0e3, +1.5e4, 0, +4.0e3)
b02 = Box3d(+0.8e4, +1.8e4 , +0.4e4, +1.4e4, 0, +4.0e3)
bt = Arr2d(xp.bt.extract(extent=b01.xy).arr, extent=np.array(b01.xy)/1000)
recs_ids = xp.recs.id
recs_x = xp.recs.gx
recs_y = xp.recs.gy
srcs_x = xp.srcs.sx
srcs_y = xp.srcs.sy
recs_x, recs_y, srcs_x, srcs_y = [i/1000 for i in [recs_x, recs_y, srcs_x, srcs_y]]
```

```python
# fig = figure(30,12)
fig = figure(15,6)
xp.plot_acq(fig=fig, extent=b01.xy, box=0, markersize=10, cmap='cmo.deep')
xp._plot_box(b02.extent[:-1], lw=3, c='w')
xp._plot_box(b01.extent[:-1], lw=8)
_ = plt.legend(prop={'size': 16}, ncol=2)
# plt.savefig(path_pap + 'acq_geom.pdf') #, dpi=300)
```

# V. 02
Arial font (set in .mplstyle file)

```python
# plot 
plt.style.reload_library()
plt.style.use(['default', 'paper2022kolumbo']) # this sets Arial font
figure(6,6)
bt.plot(mode='shade', cmap='cmo.deep', label='seabed depth [m b.s.l.]')
bt.plot(mode='cr', colors='k', show_cntr_labels=True, fontsize=5)
shotsize, shotcolor = 2, 'w'
plt.plot(srcs_x, srcs_y, linestyle='', marker='.', label='shot',
         markersize=1, markerfacecolor=shotcolor, markeredgecolor=shotcolor)
plt.plot(recs_x, recs_y, marker='o', linestyle='', markersize=shotsize*3, 
               markeredgecolor='k', markeredgewidth=1, markerfacecolor='w')
shift = 0.5
for ID, x, y in zip(recs_ids, recs_x, recs_y):
    xytext = (x-shift, y+shift)
    plt.annotate(s=str(ID), xy=(x,y), xytext=xytext, clip_on=True,
                bbox={'facecolor': 'w', 'edgecolor': 'w', 'alpha': .5, })
plt.xlabel('x [km]')
plt.ylabel('y [km]')
# plt.xticks(range(10,24,5))
# plt.yticks(range(0,12,5))
plt.xlim(np.array(b01.xy[0])/1000)
plt.ylim(np.array(b01.xy[1])/1000)
plt.gca().set_aspect('equal')
plt.savefig('fig01b.pdf', dpi=300)
plt.show()
```
